```python
import pickle
import matplotlib.pyplot as plt
```

# Basic Scatter Plots

```python
df = pickle.load(open("music_backup.p", "rb"))
```

```python
df
```

```python
for x in range(2, len(df.columns)):
    for y in range(x + 1, len(df.columns)):
        plt.scatter(df.iloc[:,x], df.iloc[:,y])
        plt.xlabel(df.columns[x])
        plt.ylabel(df.columns[y])
        plt.show()
```
